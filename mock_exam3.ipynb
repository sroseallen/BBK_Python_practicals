# Q3a (15 points)

The cells are divided in two groups, "WT" and "GM". Compute the mean probability for contacts within each group.

```python
def filter_mean(df, reg: str)-> float:
    df = df.filter(regex=reg, axis=1).filter(regex=reg, axis=0)
    mean = df.to_numpy().mean()
    return mean

WT = filter_mean(df, 'WT')
GM = filter_mean(df, 'GM')
print(WT, GM)
```

# Q3b (15 points)

Are the distribution of values significantly different?

Write code to test this assumption, and return the p-value of an appropriate test. (Tip: can you assume the numbers are normally distributed?)

```python
def filter_set(df, reg:str)->float:
    df = df.filter(regex=reg, axis=1).filter(regex=reg, axis=0)
    df = df.to_numpy().flatten() # needed to convert sample observations from dataframe to 1D array - allows statistical comparison tests below
    return df

WT = filter_set(df, 'WT')
GM = filter_set(df, 'GM')

import scipy.stats as sc

# test for normal distribution
def normal_test(df):
    a, p = sc.normaltest(df)
    if p < 0.001:
        return True
    else:
        return False

# decision between pearson and spearman test for statistical significance
def significance_p(df1, df2, df1_normal=True, df2_normal=True):
    if df1_normal==True and df2_normal==True: #if parametric
        pearson=sc.pearsonr(df1, df2)[1]
        print("pearson used:", pearson)
        return pearson
    else: # non-parametric test needed, at least one dataset not normally distributed
        spearman=sc.spearmanr(df1, df2)[1]
        print("spearman used")
        return spearman

# print significance result
result = significance_p(WT, GM, normal_test(WT), normal_test(GM))
if result > abs(0.5):
    print("significant difference between WT and GM datasets")
else:
    print("non-significant difference")
```
