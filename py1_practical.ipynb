**Hint:** The formula for calculating the circumference of a circle is 2πr (where r is the radius). The formula for calculating the area of a circle is πr<sup>2</sup>. (As you don’t know how to access a value for π yet – covered in a later lecture – I've defined it for you.)

## C) Something A Bit More Challenging…

**1. Ancestors**

Suppose you were born in the 1990s and that a new generation occurs every 25 years. In this case, 16 of your ancestors would have been born in the 1890s (2 parents and 4 generations gives 24 = 16). Write a script that allows you to enter any year and estimate the number of your ancestors born around that time, to the nearest generation. (Note: this calculation takes no account of *pedigree collapse*!) 

```python
birth_year = int(input("Enter your birth year: "))
ancestor_year = int(input("Enter the year to predict ancestors for: "))
generation_time = 25

generations_passed = round(int(birth_year - ancestor_year) / generation_time)
ancestors_elapsed = 2 ** generations_passed

if ancestors_elapsed <= 1:
    print ("You are the most recent ancestor!")
else:
    print("Estimated number of ancestors born around that year: ", ancestors_elapsed)
```

**2. Days, hours, minutes, seconds**

Write a script that reads in a positive integer representing a time measured in seconds and prints out the same time broken down into days, hours, minutes and seconds.

```python
time_in_seconds = abs(int(input ("Enter a time (in seconds):")))

secs_per_min = 60
secs_per_hour = secs_per_min * 60
secs_per_day = secs_per_hour * 24

# days
days = time_in_seconds // secs_per_day
print('Days: ', days)

# hours
remaining_secs = time_in_seconds % secs_per_day
hours = remaining_secs // secs_per_hour
print('Hours: ', hours)

# minutes
remaining_secs = remaining_secs % secs_per_hour
minutes = remaining_secs // secs_per_min
print('Minutes: ', minutes)

# seconds
remaining_secs = remaining_secs % secs_per_min
print('Seconds: ', remaining_secs)
```
