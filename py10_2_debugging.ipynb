## Problem 3: glycine-rich position

This code should find the position in an alignment that has the highest number of glycines

```python
from Bio import AlignIO
import numpy as np

def find_max_aa(filename: str ="../data/hAPP.clustal", extension: str ="clustal", aa: str ="G") -> int:
    """ This function reads in an alignment using AlignIO, and returns the position in the alignment
    with the greatest number of a given aminoacid (aa).
    Returns the first occurrence in case of multiple occurrences of the maximum value.
    """
    # N.B.: for the debugging notebooks, note that the extension should match the filetype used in AlignIO.read
    hAPP_alignment = AlignIO.read(filename, extension) # filename should not be in " ".
    # find the alignment length
    N = len(hAPP_alignment[0])
    # This will return a list of the number of glycines, with one value per position in the alignment
    aa_counts = [hAPP_alignment[:, i].count(aa) for i in range(N)]
    # convert to numpy array to use idxmax
    aa_counts = np.array(aa_counts)
    return np.argmax(aa_counts)

print(find_max_aa(aa='G'))
```

## Problem 4: scipy
This code should take a 1000 pseudo-random numbers from a laplace distribution, then compare it to the theoretical distribution

```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt

# we use the Laplace distribution
distribution = st.laplace
# sample from it
laplace_sample = distribution.rvs(size=1000)
# define the min and max for the theoretical distribution
x_min = distribution.ppf(0.01)
x_max = distribution.ppf(0.99)
# get evenly spaced points between the min and max
x = np.linspace(x_min, x_max)
# we get the theoretical distribution
laplace_theoretical = distribution.pdf(x)

# and now we plot it!
fig, ax = plt.subplots()
# histogram of the sample
ax.hist(laplace_sample, density=True, bins="fd")
# line plot for the theoretical distribution
ax.plot(x, laplace_theoretical)
ax.legend(["theoretical", "sample"])
plt.show()
```

## Problem 5: alphabet counter

This code should compute the frequencies of the standard 20 aminoacids in some sample protein sequences

```python
from Bio import SeqIO
import matplotlib.pyplot as plt

def make_count_plot(ax: plt.Axes, alphabet: [str, ], counts: [int, ]):
    """
    function that taxes a matplotlib axis object, an alphabet, and a set of counts,
    then display them using a bar plot

    """
    # from 0 to N, we will need it to make the plots
    x = range(len(alphabet))
    ax.bar(x, counts)
    # show the right ticks on the x-axis
    ax.set_xticks(x)
    # show the letters on the x-axis
    ax.set_xticklabels(alphabet)
    ax.set_xlabel("alphabet")
    ax.set_ylabel("counts")
    return


# alphabet, in this case amino-acids one letter code
alphabet = "ACDEFGHIKLMNPQRSTVWY"

# we will compute the counts for three protein sequences
protein_ids = ["Q8WZ42", "P42858", "P03437"]

protein_sequences = {}
for uniprot_id in protein_ids:
    seq = str(SeqIO.read(f"../data/{uniprot_id}.fasta", "fasta").seq)
    protein_sequences[uniprot_id] = seq

# this will create a new dictionary, with the same keys as texts,
# but with a list of counts for each text
prot_counts = {t: [protein_sequences[t].count(i) for i in alphabet] for t in protein_sequences}

# we want one plot per text
fig, axes = plt.subplots(1, len(prot_counts), figsize=(12, 5))

for t, ax in zip(protein_ids, axes):
    make_count_plot(ax, alphabet, prot_counts[t])
    ax.set_title(f"Protein: {t}")

plt.show()
```
